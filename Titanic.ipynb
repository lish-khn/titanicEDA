# Correlation heatmap

```python
data = df[['pclass','survived','age','sibsp','parch','fare']]
```

```python
sns.heatmap(data.corr(), annot=True, cmap='PiYG', fmt='.2f').set_title("Feature Correlation Heatmap")
```

Women had a significantly higher survival rate than men
Passengers from higher classes (Pclass 1) had a better survival chance
Younger passengers had a higher survival rate
The heatmap showed correlations, with Pclass being strong predictors of survival
